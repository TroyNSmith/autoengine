```python
from autoengine import Database, Geometry, Calculation, run

db = Database("autostore.db")

geo = Geometry(symbols=["O", "H", "H"], coordinates=[[0, 0, 0], [1, 0, 0], [0, 1, 0]])  # ty:ignore[invalid-argument-type]
calc = Calculation(program="crest", method="gfnff")
```

```python
from hyperqueue import Job, LocalCluster

with LocalCluster() as cluster:
    cluster.start_worker()

    client = cluster.client()
    job = Job()

    job.function(run.energy, args=("autostore.db",), kwargs={'calc': calc, 'geo': geo})

    submitted = client.submit(job)

    client.wait_for_jobs([submitted])
```
