```python
from automol import Algorithm
from autostorage import Database, IdentityRow, GeometryRow, ModelRow
from autoengine import optimization, InputProvenance

db = Database("test.db")

model = ModelRow.find_or_create(db, program="orca", method="xtb")

ident = IdentityRow.from_value("CC", algorithm=Algorithm.RDKIT_SMILES)
geo = ident.geometry()
geo_row = GeometryRow(symbols=geo.symbols, coordinates=geo.coordinates, charge=geo.charge, spin=geo.spin).save(db)

opt_row = optimization(db, geo_row, model, prov=InputProvenance())

opt_row.geometry
```
